Reading ECG Signal of single file

```python
import wfdb

# Define the file path (without the extension)
file_path = '/content/gdrive/MyDrive/Records/Database/patient001/s0010_re'

# Read the record and annotation
record = wfdb.rdrecord(file_path)

# Plot the signal

wfdb.plot_wfdb(
    record=record,
    title='ECG Signal for s0010_re',
    time_units='seconds',      # Display time in seconds
    figsize=(20, 20),           # Customize figure size
    return_fig=False           # Only plot the figure without returning it
)
```
